```python
# Suppress warnings for cleaner output

# This is often used in data analysis and machine learning notebooks to hide warnings that may not be relevant to the analysis, such as deprecation warnings or warnings about data types. However, it's important to use this with caution, as it can also hide important warnings that could indicate issues with the code or data.
import warnings
warnings.filterwarnings("ignore")

# Importing necessary libraries for data manipulation and visualization
import os
from pathlib import Path

# Data manipulation and visualization libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style and figure size
plt.rcParams["figure.figsize"] = (10, 6)
sns.set_theme(style="whitegrid")
```

# Load Dataset

This notebook first attempts to load the cleaned dataset created during the data preparation stage.
If that is not available, it falls back to the raw HR Employee Attrition dataset.


```python
preferred_paths = [
    Path("dataset/processed/hr_employee_attrition_cleaned.csv"),
    Path("dataset/raw/hr_employee_attrition.csv")
]

data_path = None
for path in preferred_paths:
    if path.exists():
        data_path = path
        break

if data_path is None:
    raise FileNotFoundError(
        "Could not find the HR Employee Attrition dataset. "
        "Expected one of: dataset/processed/hr_employee_attrition_cleaned.csv "
        "or data/raw/hr_employee_attrition.csv"
    )

df = pd.read_csv(data_path)
print("Loaded dataset from:", data_path)
print("Dataset shape:", df.shape)
df.head()
```

# 1. Attrition Distribution

This visual supports the business requirement to understand the scale of the attrition problem.


```python
attrition_counts = df["Attrition"].value_counts()
attrition_counts
```

```python
sns.countplot(data=df, x="Attrition")
plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

**Insight:** This chart shows the balance between employees who stayed and employees who left. It gives immediate visibility into the size of the attrition issue within the dataset.

**Business Recommendation:** Use this as a high-level KPI in the dashboard so stakeholders can quickly understand the overall attrition level before exploring root causes.


---
# 2. Workforce Composition

The README identifies histograms as useful for understanding workforce composition. This section explores the distribution of age and income.


```python
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

**Insight:** The age histogram shows how employees are distributed across age groups. This helps assess whether attrition patterns may be concentrated among specific workforce segments.

**Business Recommendation:** Compare age-based attrition patterns in later dashboard views to identify whether different retention strategies may be needed for early-career, mid-career, or long-tenure staff.


```python
sns.histplot(df["MonthlyIncome"], bins=20, kde=True)
plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

**Insight:** Income distribution provides context for understanding salary spread across the workforce and supports later analysis of whether lower pay is associated with higher attrition.

**Business Recommendation:** Use this alongside salary-vs-attrition visuals to identify pay bands that may require review or benchmarking.


---
# 3. Attrition by Department

This directly supports the README business requirement to analyse attrition trends across departments and roles.


```python
sns.countplot(data=df, x="Department", hue="Attrition")
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```

**Insight:** Department-level attrition highlights where turnover may be concentrated. This helps identify whether specific operational areas are under greater retention pressure.

**Business Recommendation:** Prioritise deeper investigation into departments with visibly higher attrition so HR can review workload, management practices, progression opportunities, or job design.


---
# 4. Attrition by Job Role

This supports the requirement to analyse attrition trends across roles.


```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y="JobRole", hue="Attrition", order=df["JobRole"].value_counts().index)
plt.title("Attrition by Job Role")
plt.xlabel("Number of Employees")
plt.ylabel("Job Role")
plt.tight_layout()
plt.show()
```

**Insight:** Role-level attrition can reveal more specific pressure points than department-level analysis alone. It helps identify whether particular job types are more vulnerable to turnover.

**Business Recommendation:** Focus targeted retention plans on high-risk roles, including role-specific development pathways, workload reviews, or targeted compensation analysis.


---
# 5. Compensation Impact

The README maps box plots to the business requirement of evaluating the impact of compensation on attrition.


```python
sns.boxplot(data=df, x="Attrition", y="MonthlyIncome")
plt.title("Monthly Income by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.tight_layout()
plt.show()
```

**Insight (H2):** This box plot supports the hypothesis that lower income is associated with higher attrition. The relationship may not be the only factor, but compensation appears to influence turnover risk.

**Business Recommendation:** Review salary competitiveness for groups showing lower income and higher attrition, especially where pay interacts with satisfaction or tenure.


---
# 6. Job Satisfaction and Attrition

This section reflects the README hypothesis that low job satisfaction increases attrition.


```python
sns.countplot(data=df, x="JobSatisfaction", hue="Attrition")
plt.title("Job Satisfaction vs Attrition")
plt.xlabel("Job Satisfaction")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

**Insight (H1):** Lower job satisfaction levels are associated with higher attrition, supporting the README hypothesis that dissatisfaction is a key driver of employee turnover.

**Business Recommendation:** Use satisfaction data as an early warning indicator and strengthen feedback, recognition, and development initiatives in groups with lower satisfaction levels.


---
# 7. Work-Life Balance and Attrition

This section reflects the README hypothesis that poor work-life balance increases attrition.


```python
sns.countplot(data=df, x="WorkLifeBalance", hue="Attrition")
plt.title("Work-Life Balance vs Attrition")
plt.xlabel("Work-Life Balance")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
```

**Insight (H3):** Poorer work-life balance is linked to higher attrition, supporting the hypothesis that employee wellbeing and flexibility influence retention.

**Business Recommendation:** Consider flexible working arrangements, wellbeing support, and workload management interventions for employee groups showing weaker work-life balance.


---
# 8. Years at Company

Tenure helps show whether attrition is concentrated at particular stages of the employee lifecycle.


```python
sns.boxplot(data=df, x="Attrition", y="YearsAtCompany")
plt.title("Years at Company by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Years at Company")
plt.tight_layout()
plt.show()
```

**Insight:** Tenure patterns help identify whether attrition risk is concentrated among newer employees, established staff, or longer-serving employees.

**Business Recommendation:** Use tenure-based segmentation in the dashboard to target onboarding, career development, or retention interventions at the right stage of the employee journey.


---
# 9. Correlation Heatmap

The README maps heatmaps to identifying key relationships and drivers of attrition.


```python
numeric_df = df.select_dtypes(include=[np.number])

corr = numeric_df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

**Insight:** The heatmap shows relationships between numerical variables and helps highlight which features may move together. This supports feature analysis and later modelling decisions.

**Business Recommendation:** Use strongly related variables as inputs for predictive modelling and decision-support views in the dashboard, while being careful not to interpret correlation as causation.
